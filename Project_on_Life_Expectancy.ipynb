```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/Carrer 247/Data Sheet/Life Expectancy Data.csv")
df.head(4)
```

```python
print("Shape of Data:", df.shape)
df.info()
df.describe()
```

```python
df.columns = df.columns.str.strip().str.replace(r"\s+", " ", regex=True).str.lower()
df.info()
```

```python
df.isnull().sum()
```

```python
print(df.isnull().mean())
```

```python
# Filling missing numeric values with the mean

df = df.fillna(df.mean(numeric_only=True))
```

```python
# Checking the missing value

print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Variables actually affecting the life expectancy?

# Correlation with Life Expectancy
corr = df.corr(numeric_only=True)["life expectancy"].sort_values()

print(corr)
```

```python
corr.tail(10).plot(kind="barh", figsize=(6,4))
plt.title("Top Predictors Affecting Life Expectancy")
plt.xlabel("Correlation")
plt.show()
```

```python
# Health Expenditure vs Life Expectancy Group
```

```python
df['group'] = df['life expectancy'].apply(lambda x: "Low (<65)" if x < 65 else "High (>=65)")

avg_expenditure = df.groupby("group")["total expenditure"].mean()
print(avg_expenditure)

# Simple bar graph
avg_expenditure.plot(kind="bar", figsize=(5,4))
plt.title("Health Expenditure vs Life Expectancy Group")
plt.ylabel("Average Health Expenditure")
plt.show()
```

```python
# Countries with life expectancy below 65 usually spend less on healthcare.
# So increasing healthcare expenditure can help improve life expectancy.
```

```python
print(df[['infant deaths', 'adult mortality','life expectancy']].corr())
```

```python
data = df[["infant deaths", "adult mortality", "life expectancy"]]

# Create heatmap
plt.figure(figsize=(5,4))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")

plt.title("Heatmap: Mortality Rates vs Life Expectancy")
plt.show()
```

```python
# Adult mortality is the strongest negative factor impacting life expectancy in this dataset.
# Reducing adult mortality (through better healthcare, disease prevention, and safety measures),
#    would significantly improve life expectancy across countries.
# Besides this we should also not ignore the mortality rate of infant, It should be taken to zero.
```

```python
corr = df[["life expectancy", "bmi", "thinness 1-19 years", "thinness 5-9 years", "alcohol"]].corr()
print(corr)
```

```python
# Select lifestyle-related columns
lifestyle_cols = [
    "life expectancy",
    "bmi",
    "alcohol",
    "thinness 1-19 years",
    "thinness 5-9 years"
]

data = df[lifestyle_cols]

# Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")

plt.title("Heatmap of Lifestyle Factors vs Life Expectancy")
plt.show()
```

```python
# BMI â†’ Positive correlation
# Healthy weight improves life expectancy.

# Thinness â†’ Negative correlation
# (Malnutrition reduces lifespan)

# Alcohol â†’ Slight negative or weak correlation
# Heavy drinking lowers life expectancy.
```

```python
# Impact of Schooling in Life Expectancy

df["school_group"] = pd.qcut(df["schooling"], 3, labels=["Low Schooling", "Medium Schooling", "High Schooling"])

# Average life expectancy for each group
school_bar = df.groupby("school_group")["life expectancy"].mean()

# Bar chart
plt.figure(figsize=(6,4))
school_bar.plot(kind="bar")

plt.title("Impact of Schooling on Life Expectancy")
plt.ylabel("Average Life Expectancy")
plt.xlabel("Schooling Level")
plt.show()
```

```python
# Higher schooling â†’ higher life expectancy
# Education improves health, income, and lifestyle, which leads to longer life.
```

```python
df["pop_group"] = pd.qcut(df["population"], 3, labels=["Low Population", "Medium Population", "High Population"])

# Average life expectancy per group
pop_bar = df.groupby("pop_group")["life expectancy"].mean()

# Bar chart
plt.figure(figsize=(6,4))
pop_bar.plot(kind="bar")

plt.title("Life Expectancy vs Population Level")
plt.ylabel("Average Life Expectancy")
plt.xlabel("Population Group")
plt.show()
```

```python
# here we can see that, countries having low population has high life expectancy
```

```python
df["immunization_avg"] = df[["polio", "diphtheria", "hepatitis b"]].mean(axis=1)

# 3) Create three groups: Low / Medium / High immunization
df["immunization_group"] = pd.qcut(df["immunization_avg"], 3,
                                   labels=["Low Immunization", "Medium Immunization", "High Immunization"])

# average life expectancy
avg_life = df.groupby("immunization_group")["life expectancy"].mean()
print("Average life expectancy by immunization group:")
print(avg_life)

plt.figure(figsize=(6,4))
avg_life.plot(kind="bar")
plt.title("Impact of Immunization Coverage on Life Expectancy")
plt.ylabel("Average Life Expectancy")
plt.xlabel("Immunization Level")
plt.tight_layout()
plt.show()
```
